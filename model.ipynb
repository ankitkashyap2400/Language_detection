```python
import pandas as pad 
df=pad.read_csv('language.csv')
df.head()
```

```python
df.info
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(df['language'])
```

```python
pickle.dump(le, open("label_encoder.pkl", "wb"))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(1,3))
X = vectorizer.fit_transform(df['Text'])
```

```python
import pickle

pickle.dump(vectorizer, open("vectorizer.pkl", "wb"))
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
x_train,x_test ,y_train,y_test =train_test_split(X,y,test_size=.2)
```

```python
x_train
```

```python
y_test
```

```python
nb=MultinomialNB()
nb.fit(x_train,y_train)
```

```python
nb.score(x_test,y_test)
```

```python
pickle.dump(nb, open("nb_model.pkl", "wb"))
```

```python
from sklearn.svm import LinearSVC

svc = LinearSVC()
svc.fit(x_train, y_train)

```

```python
svc.predict(x_test)
svc.score(x_test,y_test)
```

```python
pickle.dump(svc, open("svc_model.pkl", "wb"))
```

```python

import numpy as np
user_input = input("Enter text: ")
text = user_input.lower()
text_vec = vectorizer.transform([text])

# SVM
pred_svc = svc.predict(text_vec)
print("SVC:", le.inverse_transform(pred_svc)[0])

# NB
pred_nb = nb.predict(text_vec)
prob_nb = np.max(nb.predict_proba(text_vec))  # confidence

print("NB :", le.inverse_transform(pred_nb)[0])
print("NB Confidence:", round(prob_nb, 3))
```
